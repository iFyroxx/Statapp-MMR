```python
clean = uniques[["gameCreation", "gameDuration", "gameId", "gameMode", "participantIdentities", "participants", "queueId"]].sort_values(by = "gameCreation")
```

```python
import time

clean["gameCreation"] = clean["gameCreation"].apply(lambda x : int(x))
clean["year"] = clean["gameCreation"].apply(lambda t : time.gmtime(t/1000)[0])
clean["month"] = clean["gameCreation"].apply(lambda t : time.gmtime(t/1000)[1])
clean["day"] = clean["gameCreation"].apply(lambda t : time.gmtime(t/1000)[2])
clean["hour"] = clean["gameCreation"].apply(lambda t : time.gmtime(t/1000)[3])
clean["minute"] = clean["gameCreation"].apply(lambda t : time.gmtime(t/1000)[4])
clean["second"] = clean["gameCreation"].apply(lambda t : time.gmtime(t/1000)[5])
clean["gameId"] = clean["gameId"].apply(lambda x : int(x))
```

```python
clean = clean[clean["gameMode"] == "CLASSIC"]
clean = clean.reset_index(drop = True)
```

```python
clean = clean[["gameCreation", "year", "month", "day", "hour", "minute", "second", "gameId", "gameMode", "participantIdentities", "participants", "queueId"]]
```

```python
clean.head(5)
```

```python
game_history_accounts = {}
for i in range(clean.shape[0]):
    participants = clean["participantIdentities"].iloc[i]
    gameID = clean["gameId"].iloc[i]
    for p in participants:
        account = p["player"]["accountId"]
        if account not in game_history_accounts.keys():
            game_history_accounts[account] = []
        game_history_accounts[account].append(int(gameID))
```

```python
# On va créer la liste des participants :

participants = []

for game in clean["participantIdentities"]:
    for participant in game:
        participants.append(participant['player']['accountId'])

participants = list(np.unique(participants))
```

```python
import random

extract = random.sample(participants, 20000)
```

```python
P = [tuple([0]*len(extract))]*len(extract)
```

```python
P = pd.DataFrame(P, index = extract, columns = extract)
```

```python
for game in clean["participantIdentities"]:
    for i in range(len(game)):
        if game[i]['player']['accountId'] in extract:
            for j in range(i+1,len(game)):
                if game[j]['player']['accountId'] in extract:
                    P[game[i]['player']['accountId']][game[j]['player']['accountId']] += 1
                    P[game[j]['player']['accountId']][game[i]['player']['accountId']] += 1
```

```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
def laplacian(M):
    D = np.sum(M, axis = 1)
    return D-M

L = laplacian(P)
```

```python
import numpy.linalg as linalg

eigenValues, eigenVectors = linalg.eig(L)
```

```python
idx = eigenValues.argsort()
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:,idx]
```

```python
for k in range(50):
    features = eigenVectors[:, :k]
    kmeans = KMeans(init="random", n_clusters=k, n_init=10, max_iter=300, random_state=42)
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    kmeans.fit(scaled_features)
    print(kmeans.labels_)
```
